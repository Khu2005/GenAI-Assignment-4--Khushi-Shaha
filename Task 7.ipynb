```python
prices = {"Mouse":500, "Keyboard":800, "Monitor":7000, "Pendrive":400, "Camera":5000}
discount_percent=int(input("Enter the discount percentage: "))
for product, price in prices.items():
    discount_amount = price * (discount_percent / 100)
    discounted_price = price - discount_amount
    print(f"{product}: Original Price: {price}, Discounted Price: {discounted_price}")
```

```python
with open ("discount_report.txt","w") as f:
    for product, price in prices.items():
        discount_amount = price * (discount_percent / 100)
        discounted_price = price - discount_amount
        f.write(f"{product}: Original Price: {price}, Discounted Price: {discounted_price}\n")
```
