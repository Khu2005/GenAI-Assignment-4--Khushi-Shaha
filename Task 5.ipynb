```python
products=[]
price=[]

for i in range(3): 
    product = input(f"Enter the Name of Product {i + 1}: ")
    products.append(product)
    p = int(input(f"Enter the price of {product}: "))
    price.append(p)
print("Products: ", products)
print("Prices: ", price)
```

```python
with open ("products.txt", "w") as f:
    for i in range(3):
        f.write(f"{products[i]}|{price[i]}\n")
```
