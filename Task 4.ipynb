```python
with open ("sales.txt") as f:
    data = sum(int(line) for line in f)
    print(f"Total sales: {data}")
```

```python
with open ("sales.txt") as f:
    data = max(int(line) for line in f)
    print(f"Maximum sales: {data}")
```

```python
with open ("sales.txt") as f:
    data = min(int(line) for line in f)
    print(f"Minimum sales: {data}")
```

```python
with open ("sales.txt") as f:
    data = [int(line) for line in f]
    print(f"Average sales: ",sum(data)/len(data))
```
